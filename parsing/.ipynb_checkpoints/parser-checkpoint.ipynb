```python
from bs4 import BeautifulSoup
import string

PATH = 'wpcd/wp/'
```

```python
index = list(string.ascii_lowercase) #list of index pages
index.append('others')
```

```python
for sub_index in [f'{name}-list.htm' for name in index]:
    print(sub_index)
```

```python
sub_index = 'a-list.htm'

with open(f'{PATH}/index/{sub_index}', encoding = 'utf8') as fp:
    soup = BeautifulSoup(fp, 'html.parser')

for link in soup.find_all('a', href = True):
    assert link['href'][:9] == '../../wp/'
    print(link['href'][9:])
```

```python
page_url = 'a/A._E._J._Collins.htm'

with open(f'{PATH}/{page_url}', encoding = 'utf8') as fp:
    page_soup = BeautifulSoup(fp, 'html.parser')

for link in page_soup.find_all('a', href = True):
    if (link['href'][:9] == '../../wp/'):
        print(link['href'][9:])
```
